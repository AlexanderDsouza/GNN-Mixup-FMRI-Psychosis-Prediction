```python
import pandas as pd
import numpy as np

# Load Excel file
file_path = '/content/drive/Shared drives/GNN/SchaeferAtlas_Rest_Results_Freq_008to09_wholebrain.xlsx'
excel_file = pd.ExcelFile(file_path)

# Dictionary to store each patient's connectivity matrix
patient_matrices = {}

# Loop through all sheets (patients)
for sheet_name in excel_file.sheet_names:
    df = pd.read_excel(file_path, sheet_name=sheet_name)

    # Exclude first column if it's labels
    numeric_data = df.iloc[:, 1:]

    # Average across rows (measurements) to get one vector per patient
    avg_vector = numeric_data.mean(axis=0)

    # Convert the vector into a symmetric connectivity matrix
    n = len(avg_vector)
    matrix = np.zeros((n, n))

    for i in range(n):
        for j in range(i, n):
            matrix[i, j] = avg_vector.iloc[j]
            matrix[j, i] = avg_vector.iloc[j]

    # Convert to DataFrame with region names as labels
    regions = df.columns[1:]
    conn_matrix = pd.DataFrame(matrix, index=regions, columns=regions)

    # Store in dictionary
    patient_matrices[sheet_name] = conn_matrix

# Example: print one patient's matrix
print(patient_matrices[excel_file.sheet_names[0]])
```
